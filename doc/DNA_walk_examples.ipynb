## Przykładowe sygnały stworzone za pomocą DNA Walk

```python
from matplotlib import pyplot as plt
```

```python
def get_walk (sequence):
    x = 0
    y = 0
    x_list = []
    y_list = []
    for s in sequence:
        x = x+1 if s=="A" else x if s=="G" or s=="T" else x-1
        y = y+1 if s=="G" else y if s=="A" or s=="C" else y-1
        x_list.append(x)
        y_list.append(y)
    return (x_list,y_list)
```

```python
sequence = "AAAGATTCTATGCCTCCGGGGAACAATGTGCCTCAGAGCTGGTATGAAGCTAAAAAGATAATGTCTAATCTGGGTCTGAA\\
CTGTGAGAAGATTCATGCATGTCCTAATGATTGTC"

(x_list,y_list) = get_walk(sequence)
plt.plot(x_list,y_list)
plt.show()
```

```python
sequence =  "CTAACTTTGACTTATTTAGGTTAGTATGACACGTGACTATATCAGAGTTAAGGATTGGATGTATGTGAATCGTGAAAAAT\\
ACACAAAGCCTGAATTCATTGAAGGAGTTGCTTCTTTCATAAAATTTGTGCGTGAACACGATAGGATCCAACATGGGATG\\
ATTCTTTGTCCATGCAGTCATTATGGGAATAGGTTGCATCGTAAGGAGGAAGATGTAGAACTTCACATTGTTAAGCATGG\\
TTTTATTAGAGATTATACAATTTGGGTATATCACAGAGAAATTCT"

(x_list,y_list) = get_walk(sequence)
plt.plot(x_list,y_list)
plt.show()
```
